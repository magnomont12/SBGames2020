```
###Imporsts
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```
XHP = np.array([[64, 1722,2268, 2673, 4652,5015,5869, 6059, 8192, 8770]])
XHP = XHP.T

XATQ = np.array([[41, 58, 118, 135, 185,313,340, 404, 457, 520]])
XATQ = XATQ.T

XDEF = np.array([[22,64, 81, 238, 242, 244,338, 510,611, 722]])
XDEF = XDEF.T

XAGI = np.array([[0,9,10, 10,22, 27,31, 43, 49,84]])
XAGI = XAGI.T

XESQV = np.array([[0,0,2, 4,6, 9,9, 11, 15, 27]])
XESQV = XESQV.T

XCRIT = np.array([[0,1,3,4,6,10,12,12,24,43]])
XCRIT = XCRIT.T
```

```
y = np.array([[1,10,20,30,40,50,60,70,80,90]])
y = y.T
```

```
# Training the Polynomial Regression model on the whole dataset
poly_reg = PolynomialFeatures(degree = 3)
X_poly = poly_reg.fit_transform(y)

##HP
reg_hp = LinearRegression()
reg_hp.fit(X_poly, XHP)

##ATQ
reg_atq = LinearRegression()
reg_atq.fit(X_poly, XATQ)

##DEF
reg_def = LinearRegression()
reg_def.fit(X_poly, XDEF)

##AGI
reg_agi = LinearRegression()
reg_agi.fit(X_poly, XAGI)

##ESQV
reg_esqv = LinearRegression()
reg_esqv.fit(X_poly, XESQV)

##CRIT
reg_crit = LinearRegression()
reg_crit.fit(X_poly, XCRIT)
```

```
listLevel = []
for a in range(1, 99):
  listLevel.append([a])
```

```
reg_hp.predict(poly_reg.fit_transform(listLevel))## Variável que contem valores previstos para cada level
plt.scatter(y, XHP, color = 'red')
plt.plot(y, reg_hp.predict(poly_reg.fit_transform(y)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
